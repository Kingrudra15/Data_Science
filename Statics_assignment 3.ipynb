```python
Q1: Probability Mass Function (PMF) and Probability Density Function (PDF)

Probability Mass Function (PMF): PMF is a function that gives the probability of discrete random variables taking specific values.
It is often denoted as P(X = x), where X is the random variable and x is a specific value.

Example: Consider a fair six-sided die. The PMF for the possible outcomes (1, 2, 3, 4, 5, 6) is P(X = 1) = 1/6, P(X = 2) = 1/6,
..., P(X = 6) = 1/6.

Probability Density Function (PDF): PDF is a function that describes the likelihood of continuous random variables falling within 
a particular range. It is represented as f(x), and the probability of the variable falling within a range is given by the integral 
of the PDF over that range.

Example: The normal distribution has a PDF given by the bell-shaped curve formula. For a normal distribution with mean μ and 
standard deviation σ, the PDF is f(x) = (1 / (σ√(2π))) * e^(-(x-μ)²/(2σ²)).
```

Q2: Cumulative Density Function (CDF)

The Cumulative Density Function (CDF) gives the probability that a random variable takes on a value less than or equal to a given point. It is denoted as F(x) for a random variable X.

Example: For the normal distribution, the CDF is often denoted as Φ(x) and calculated using tables or software. For instance, Φ(1) represents the probability that a standard normal random variable is less than or equal to 1.

Importance: CDF is useful for determining the probability of a random variable falling within a certain range.

Q3: Normal Distribution

The normal distribution is used in situations where data tends to cluster around a central value with symmetrically decreasing probability as values move away from the mean. Examples include heights, weights, and errors in measurements.

Parameters: The normal distribution is characterized by two parameters, mean (μ) and standard deviation (σ). The mean locates the center, and the standard deviation controls the spread.

Q4: Importance of Normal Distribution

Real-life Examples:
Heights of a population.
IQ scores in a large population.
Errors in measurements.
Blood pressure in adults.

Q5: Bernoulli Distribution

Bernoulli Distribution: It models a binary outcome, where an experiment results in either success (1) or failure (0), with a fixed probability of success (p) and failure (1-p).

Example: A single coin flip, where success is getting heads (1) and failure is getting tails (0).

Difference: Bernoulli Distribution deals with a single trial, while Binomial Distribution deals with the number of successes in a fixed number of independent Bernoulli trials.

Q7: Uniform Distribution

Uniform Distribution: All outcomes are equally likely over a given range. The probability density function is constant within the range.

Example: Rolling a fair six-sided die. Each outcome (1, 2, 3, 4, 5, 6) has an equal probability of 1/6.

Q8: Z Score

Z Score: It measures how many standard deviations a data point is from the mean. It is calculated as Z = (X - μ) / σ.

Importance: Z scores help standardize and compare data from different distributions, providing a common scale for analysis.

Q9: Central Limit Theorem

Central Limit Theorem (CLT): It states that, regardless of the shape of the original distribution, the sampling distribution of the sample mean approaches a normal distribution as the sample size increases.

Significance: It allows us to make statistical inferences about a population mean based on a sample.

Q10: Assumptions of Central Limit Theorem

1. Independence: The sampled observations must be independent.
2. Sample Size: The sample size should be sufficiently large for the CLT to apply.
3. Population Distribution: The shape of the population distribution doesn't matter if the sample size is large enough.
