```python
Q1. Explain GET and POST methods.

GET Method:
The GET method is used to request data from a specified resource.
Parameters are passed in the URL.
It is less secure for passing sensitive information, as data is visible in the URL.
GET requests are idempotent, meaning making the same request multiple times has the same effect as making it once.

POST Method:
The POST method is used to submit data to be processed to a specified resource.
Parameters are passed in the request body, not in the URL.
It is more secure for passing sensitive information, as data is not visible in the URL.
POST requests are not idempotent, meaning making the same request multiple times may have different effects.
```

```python
Q2: Why is request used in Flask?

The request object in Flask is used to access the data that the client (browser) sends to the server.
It allows the server to handle incoming data from HTML forms or other sources, enabling the extraction 
of form data, file uploads, and more
```

```python
Q3. Why is redirect() used in Flask?


In Flask, the redirect() function is used to redirect the client to a different URL. 
It is commonly used in scenarios where you want to direct the user to another route or an external URL 
after processing a request. For example, after a form submission, you might want to redirect the user to
a thank-you page.
```

```python
Q4. What are templates in Flask? Why is the render_template() function used?
Templates in Flask:
Templates in Flask are used to generate dynamic HTML content. They allow you to separate the HTML 
structure from the Python code in your application, promoting a clean and modular design. Flask uses 
the Jinja2 templating engine for rendering templates.

render_template() Function:
The render_template() function in Flask is used to render HTML templates. It takes the name of the template 
file as an argument, along with optional variables that can be passed to the template.
```
