Q1: Estimation Statistics

Point Estimate: A single value used to estimate the population parameter. Example: Sample mean (x̄) as an estimate of the population mean (μ).

Interval Estimate: A range of values used to estimate the population parameter. Example: Confidence interval for the population mean.

Q2. Write a Python function to estimate the population mean using a sample mean and standard
deviation.

```python
import math

def estimate_population_mean(sample_mean, sample_std, sample_size):
    margin_of_error = 1.96 * (sample_std / math.sqrt(sample_size))  # 1.96 corresponds to 95% confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error
    return lower_bound, upper_bound

# Example usage:
sample_mean = 60
sample_std = 8
sample_size = 25
confidence_interval = estimate_population_mean(sample_mean, sample_std, sample_size)
print("95% Confidence Interval:", confidence_interval)
```

Q8: P-Value in Hypothesis Testing

Definition: The p-value is the probability of obtaining a result as extreme or more extreme than the observed result under the null hypothesis.

Significance: If p-value < α, there is evidence to reject the null hypothesis.

Q9. Generate a Student's t-distribution plot using Python's matplotlib library, with the degrees of freedom
parameter set to 10.

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats

df = 10  # degrees of freedom
x = np.linspace(stats.t.ppf(0.01, df), stats.t.ppf(0.99, df), 100)
plt.plot(x, stats.t.pdf(x, df), label='t-distribution (df=10)')
plt.legend()
plt.title("Student's t-Distribution")
plt.show()
```

Q10. Write a Python program to calculate the two-sample t-test for independent samples, given two
random samples of equal size and a null hypothesis that the population means are equal.


```python
import scipy.stats as stats

def two_sample_t_test(sample1, sample2):
    statistic, p_value = stats.ttest_ind(sample1, sample2)
    return statistic, p_value

# Example usage:
sample1 = [20, 25, 30, 35, 40]
sample2 = [18, 22, 28, 32, 38]
statistic, p_value = two_sample_t_test(sample1, sample2)
print("Test Statistic:", statistic)
print("P-Value:", p_value)
```

Q13. A coffee shop owner wants to estimate the average daily revenue for their shop. They take a random
sample of 50 days and find the sample mean revenue to be $500 with a standard deviation of $50.
Estimate the population mean revenue with a 95% confidence interval.

```python
import scipy.stats as stats

sample_mean = 500
sample_std = 50
sample_size = 50

confidence_interval = stats.norm.interval(0.95, loc=sample_mean, scale=sample_std / (sample_size**0.5))
print("95% Confidence Interval:", confidence_interval)
```

Q14. A researcher hypothesizes that a new drug will decrease blood pressure by 10 mmHg. They conduct a
clinical trial with 100 patients and find that the sample mean decrease in blood pressure is 8 mmHg with a
standard deviation of 3 mmHg. Test the hypothesis with a significance level of 0.05.

```python
import numpy as np

# Generate random sample data for blood pressure decrease
blood_pressure_data = np.random.normal(loc=8, scale=3, size=100)

# Perform a hypothesis test
t_statistic, p_value = stats.ttest_1samp(blood_pressure_data, 10)

# Check the p-value to determine significance
if p_value < 0.05:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")

```
